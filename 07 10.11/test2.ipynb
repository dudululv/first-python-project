```python
students = {}

def add_score():
    while True:
        name = input("请输入学生姓名（输入 'q' 结束）：")
        if name == 'q':
            break
        if name in students:
            print("该学生已存在成绩！")
            continue
        score = input("请输入学生成绩：")
        if not score.isdigit():
            print("成绩必须是数字！")
            continue
        
        students[name] = int(score)
        print(f"已添加 {name} 的成绩：{score}")

def modify_score():
    name = input("请输入要修改成绩的学生姓名：")
    if name in students:
        new_score = input("请输入新的成绩：")
        if not new_score.isdigit():
            print("成绩必须是数字！")
            return
        students[name] = int(new_score)
        print(f"已修改 {name} 的成绩为：{new_score}")
    else:
        print("未找到该学生！")

def delete_score():
    name = input("请输入要删除的学生姓名：")
    if name in students:
        del students[name]
        print(f"已删除 {name} 的成绩。")
    else:
        print("未找到该学生！")

def main():
    while True:
        print("\n===== 学生成绩管理系统 =====")
        print("1. 添加学生成绩")
        print("2. 修改学生成绩")
        print("3. 删除学生成绩")
        print("4. 显示全部成绩")
        print("0. 退出系统")

        choice = input("请选择功能（1-0）：")

        if choice == '1':
            add_score()
        elif choice == '2':
            modify_score()
        elif choice == '3':
            delete_score()
        elif choice == '4':
            if students:
                print("\n--- 当前成绩列表 ---")
                for name, score in students.items():
                    print(f"{name}: {score}")
            else:
                print("暂无成绩数据。")
        elif choice == '0':
            print("系统已退出，再见！")
            break
        else:
            print("输入无效，请重新选择。")

main()
```

```python
# 定义一个空列表，用来存储所有成绩
grades = []

# 循环输入3个组的成绩
for group in range(1, 4):
    group_grades = []  # 每组的成绩
    print(f"请输入第 {group} 组的 3 个 Python 成绩：")
    for student in range(1, 4):
        while True:
            try:
                score = float(input(f"  学生 {student} 的成绩: "))
                if 0 <= score <= 100:
                    group_grades.append(score)
                    break
                else:
                    print("成绩必须在0~100之间，请重新输入。")
            except ValueError:
                print("请输入有效数字！")
    grades.append(group_grades)

# 打印成绩
print("\n各组成绩如下：")
for i, group_grades in enumerate(grades, start=1):
    print(f"第 {i} 组: {group_grades}")
```

```python
# 定义一个空列表，用来存储所有成绩
grades = []

# 循环输入3个组的成绩
for group in range(1, 4):
    group_grades = []  # 每组的成绩
    print(f"请输入第 {group} 组的 3 个 Python 成绩：")
    for student in range(1, 4):
        while True:
            try:
                score = float(input(f"  学生 {student} 的成绩: "))
                if 0 <= score <= 100:
                    group_grades.append(score)
                    break
                else:
                    print("成绩必须在0~100之间，请重新输入。")
            except ValueError:
                print("请输入有效数字！")
    grades.append(group_grades)

# 打印成绩
print("\n各组成绩如下：")
for i, group_grades in enumerate(grades, start=1):
    print(f"第 {i} 组: {group_grades}")

# 计算每组的统计数据
for i, group_grades in enumerate(grades, start=1):
    total = sum(group_grades)
    highest = max(group_grades)
    lowest = min(group_grades)
    average = total / len(group_grades)
    print(f"\n第 {i} 组统计数据：")
    print(f"  总成绩: {total}")
    print(f"  最高成绩: {highest}")
    print(f"  最低成绩: {lowest}")
    print(f"  平均成绩: {average:.2f}")

# 计算整个班级的统计数据
all_scores = [score for group in grades for score in group]
total_class = sum(all_scores)
highest_class = max(all_scores)
lowest_class = min(all_scores)
average_class = total_class / len(all_scores)

print("\n班级总体统计数据：")
print(f"  总成绩: {total_class}")
print(f"  最高成绩: {highest_class}")
print(f"  最低成绩: {lowest_class}")
print(f"  平均成绩: {average_class:.2f}")
```

```python
import random
from collections import Counter

# 随机生成50个1~50之间的整数
numbers = [random.randint(1, 50) for _ in range(50)]
print("生成的随机数列表：")
print(numbers)

# 统计每个数出现的次数
count = Counter(numbers)

print("\n每个数出现的次数：")
# 按数字顺序打印
for num in range(1, 51):
    print(f"{num}: {count.get(num, 0)}")
```

```python
s = '100,23,5007,5447'

# 1. 拆分字符串为列表
numbers = s.split(',')

# 2. 转为整数排序
numbers_sorted = sorted(numbers, key=int)

# 3. 重新拼接成字符串
sorted_str = ','.join(numbers_sorted)

print(sorted_str)
```

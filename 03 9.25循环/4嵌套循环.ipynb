```python
for i in range(1,10):
    for j in range(1,i+1):
        print(f'{j}*{i}={j*i}',end='\t')
    else:
        print()
```

```python
#不相等的三位数
s=0
for i in range(1,10):
    for j in range(10):
        for k in range(10):
            if i!=j and i!=k and j!=k:
                print(f'{i}{j}{k}')
                s+=1
print(s)
```

```python
#百元百鸡公5母3小0.3
for i in range (21):
    for j in range(34):
        k=100-i-j
        if i*5+j*3+k//3==100 and k%3==0:
            print(f'公{i}母{j}小{k}')
```

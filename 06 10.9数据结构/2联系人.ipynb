```python
msg='''1添加联系人
2查看联系人
3修改联系人
4删除联系人
0退出程序
请选择：'''
names=['张三','李四']
phones=['13112341234','13212341234']
while 1:
#输入0-4
    while 1:
        key=input(msg)
        if key.isdigit():
            key=int(key)
            if 0<=key<=4:
                break
            else:
                print('输入不正确，请重新输入（0-4）')
                
        else:
            print('输入不正确，请重新输入（0-4）')
            
#根据输入判断
    if key==0:
        print('欢迎下次再来')
        break
    elif key==1:
        print('1添加联系人')
        name=input('请输入姓名：')
        phone=input('请输入手机号：')
        names.append(name)
        phones.append(phone)
        print('添加成功')
        break
    elif key==2:
        while 1:
            k=input('1查看联系人/2查询单个联系人：')
            if k.isdigit():
                k=int(k)
                if 1<=k<=2:
                    break
                else:
                    print('输入不正确，请重新输入（1-2）')
            else:
                print('输入不正确，请重新输入（1-2）')
        if k==1:
            for i in range(0,len(names)):
                print(f'{i+1}.{names[i]}{phones[i]}')
        else:
            name=input('请输入要查询的联系人姓名：')
            if name in names:
                print(f'{name}{phones[names.index(name)]}')
            else:
                print('联系人不存在')
        break
    elif key==3:
        print('3修改联系人')
        name=input('请输入要修改的联系人：')
        if name in names:        
            phone=input('请输入新手机号：')
            i=names.index(name)
            phones[i]=phone
            print(f'{name}:{phone}修改成功')
            
        else:
             print(f'{name}不存在')  
             break         
    else:
        print('4删除联系人')
        name=input('请输入要删除的联系人：')
        if name in names:
            i=names.index(name)
            names.remove(names[i])
            phones.remove(phones[i])
            print('删除成功')
            
        else:
            print(f'{name}不存在')     
            break      

```

```python
msg='''1添加联系人
2查看联系人
3修改联系人
4删除联系人
0退出程序
请选择：'''
info={'张三':1234,'李四':5678}
while 1:
    num=input(msg)
    if num=='0':
        break
    elif num=='1':
        name=input('请输入联系人姓名：')
        if name in info:
            print(f'{name}已存在')
            break
        else:
            phone=input('请输入联系人电话：')
            info[name]=phone
            print(f'{name}:{phone}添加成功')
    elif num=='2':
        num2=input('1查看所有联系人\n2查看指定联系人：')
        if num2=='1':
            print(info)
        elif num2=='2':
            name=input('请输入要查看的联系人姓名：')
            if name in info:
                print(f'{name}:{info[name]}')
            else:
                print(f'{name}不存在')
                break 
        else:
            print('输入错误，请重新输入')
        break
    elif num=='3':
        name=input('请输入要修改的联系人姓名：')
        if name in info:
            phone=input('请输入联系人电话：')
            info[name]=phone
            print(f'{name}:{phone}修改成功')
        else:
            print(f'{name}不存在')
            break
    elif num=='4':
        name=input('请输入要删除的联系人姓名：')
        if name in info:
            del info[name]
            print(f'{name}删除成功')
        else:
            print(f'{name}不存在')
            break
    else:
        print('输入错误，请重新输入')
```

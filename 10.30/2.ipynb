```python
import numpy as np
import pandas as pd

df = pd.read_csv('../data/player.csv', index_col=0)
df = df.assign(height_class=np.select([
    df.height >= 185,
    (df.height >= 165) & (df.height < 185)
], [2, 1], 0))
df.set_index('height_class', append=True, inplace=True)
#这个或
df = df.reset_index('player_api_id')
df.set_index('player_api_id', append=True, inplace=True)
df
```

```python
#高维索引的操作
np.random.seed(0)
df = pd.DataFrame(np.random.randint(0,20,(4,4)),
                  index=[['上半年','上半年','下半年','下半年'],['A组','B组','A组','B组']],
                  columns=[['一区','一区','二区','二区'],['苹果','橘子','苹果','橘子']])
df
# 列操作:df[高维索引，低维索引]
# 获得一区的数据
df['一区']
```

```python
# 获得一区、橘子的数据
df['一区','橘子']
```

```python
#行操作:df.loc[高维索引，低维索引]
# 获得上半年数据
df.loc['上半年']
```

```python
#获得上半年A组数据
df.loc['上半年','A组']
```

```python
# 使用标签索引列表获得数据
df.loc[['上半年','下半年']]
```

```python
# 使用布尔列表获取数据
df.loc[[True,True,False,False]]
# 使用切片获取数据
df.loc['上半年':'下半年']
```

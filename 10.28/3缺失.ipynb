```python
#实现缺失值处理丢弃
df=pd.read_csv('../data/player_attributes.csv')
df.head(2)
#使用dropna（）实现pLayer_attributes球员评价数据表下的缺失值丢弃
#只要有任何－一个记录有缺失，则删除整行
df.dropna().isnull() .any().sum()
#只要有任何一个记录有缺失，则删除整列
df.dropna(axis=1)
#只考虑vision列和sLiding_tackle列，只要有任何一个记录缺失，则删除该行
df.dropna(subset=['vision','sliding_tackle'])
#使用notnuLl（）方法过滤缺失值数据
# 过滤掉overall_rating为空的行记录
df[df.overall_rating.notnull()]
```

```python
#实现缺失值处理填充
df = pd.read_csv('../data/player_attributes.csv')
# 使用fillna（)实现player_attributes表缺失数据的填充
# 为方便查看结果，取overallrating为空的全部记录和部分随机抽取的记录
a=df[df.overall_rating.isna()]
a.sample(10)
# 将缺失的数据统一填充为-1
a.fillna(-1)
# 将overallrating的值设置为该列的均值
df.overall_rating.mean()
a.fillna(df.overall_rating.mean() .round(2))
# 将overall_rating的值设置为该列上一个有效值
df.overall_rating.fillna(method='ffill')
#使用行列混合操作填充指定列
# 将overall_rating为空的数据填充为对应potential列的均值
df.potential.mean()
df.overall_rating=df.overall_rating.fillna(df.potential.mean())
df.query('overall_rating==@df.potential.mean()')
```

```python
#df.dtypes
#足球数据表集中，出现了负分数或者超过100的分数可以对其进行修正
#<0的分数设置为0
#>100的分数设置为100
#使用两种方法实现异常值的恢复
df = pd.read_csv('../data/player_attributes.csv')
df.dtypes
df.loc[df.overall_rating<50,'overall_rating']=-50
df.loc[df.overall_rating>50,'overall_rating']=101
for col in df.columns:
#'object'==df[col].dtype
#过滤掉列名带id的
    if 'id' not in col and'object' != df[col].dtype:
        df.loc[df[col]<0,col].size
        df.loc[df[col]>100,col].size
        df.loc[df[col]<0,col]=0
        df.loc[df[col]>100,col]=100
```

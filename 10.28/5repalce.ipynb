```python
# replace()函数的应用
import pandas as pd
# 实现player数据集中身高数据的替换
df = pd.read_csv('../data/player.csv')
df.head(2)

# 1. 将170.18的数据均替换为170.19
df.height.replace(170.18, 170.19)

# 2. 将170.18,180.34,182.88这几种身高均替换为180.00
df.height.replace([170.18, 180.34, 182.88], 180.00)

# 3. 利用列表实现以下数据替换
# 170.18->170.00
# 180.34->180.00  
# 182.88->182.00
df.height.replace([170.18, 180.34, 182.88], [170.00, 180.00, 182.00])

# 4. 利用字典实现上述数据替换（修正语法错误）
df.height.replace({170.18: 170.00, 180.34: 180.00, 182.88: 182.00})
df
```

```python
import pandas as pd
import re

df = pd.read_csv('input.csv', encoding='utf-8')
print("原始数据:")
print(df)
print("\n" + "="*50 + "\n")

# 第一题：打印Field_01每种值出现的次数
print("1. Field_01每种值出现的次数:")
field_01_counts = df['Field_01'].value_counts()
print(field_01_counts)
print()
```

```python
# 第二题：清洗Field_02，仅保留汉字
def clean_chinese(text):
    # 使用正则表达式匹配汉字，删除字母、数字、空白符和换行符
    if pd.isna(text):
        return text
    # 匹配所有汉字
    chinese_chars = re.findall(r'[\u4e00-\u9fff]', str(text))
    return ''.join(chinese_chars) #''.join() 将列表连接成字符串

df['Field_02'] = df['Field_02'].apply(clean_chinese) #对每一行应用clean_chinese函数
print("2. 清洗后的Field_02:")
print(df['Field_02'])
print()
```

```python
# 第三题：清洗Field_03，仅保留数字
def clean_numbers(text):
    if pd.isna(text):
        return text
    # 匹配所有数字
    numbers = re.findall(r'\d', str(text))
    return ''.join(numbers)

df['Field_03'] = df['Field_03'].apply(clean_numbers)
print("3. 清洗后的Field_03:")
print(df['Field_03'])
print()
```

```python
# 第四题：给Field_05赋值，值为Field_04中"s"出现的次数
def count_s_occurrences(text):
    if pd.isna(text):
        return 0
    return str(text).count('s')

df['Field_05'] = df['Field_04'].apply(count_s_occurrences)
print("4. Field_05 (Field_04中s出现的次数):")
print(df['Field_05'])
print()
```

```python
# 第五题：删除序号为4的一行
df = df[df['序号'] != 4]
print("5. 删除序号4后的数据:")
print(df)
print()
```

```python
# 第六题：填充Field_06列中的空值为"W"
df['Field_06'] = df['Field_06'].fillna('W')
print("6. 填充空值后的Field_06:")
print(df['Field_06'].head(10))
print()
```

```python
# 保存最终结果
df.to_csv('inputr.csv', index=False, encoding='utf-8-sig')
print("最终结果已保存为 inputr.csv")
print("\n最终数据:")
print(df)
```

```python
# 1. 选择订单日期、销售额和利润这三列
selected_columns = df2[['订单日期', '销售额', '利润']]

# 2. 检查并填充销售额列的缺失值
if df2['销售额'].isnull().sum() > 0:
    df2['销售额'] = df2['销售额'].fillna(0)

# 3. 检查并填充数量列的缺失值
if df2['数量'].isnull().sum() > 0:
    df2['数量'] = df2['数量'].fillna(0)

# 4. 将省/自治区列重命名为省份
df2 = df2.rename(columns={'省/自治区': '省份'})

# 5. 将装运模式列中的"二级"替换为"二级配送"
df2['装运模式'] = df2['装运模式'].replace('二级', '二级配送')

# 显示结果
print("选择的列：")
print(selected_columns.head())
print("\n销售额缺失值数量：", df2['销售额'].isnull().sum())
print("数量缺失值数量：", df2['数量'].isnull().sum())
print("\n列名已更新：")
print(df2.columns.tolist())
print("\n装运模式唯一值：")
print(df2['装运模式'].unique())
```

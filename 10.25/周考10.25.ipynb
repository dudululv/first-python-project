```python
#第一题: 读取表格"./input.csv"。请完成下面题目
# 1.请打印出，字段"Field 01"，每种值出现的次数。
# 2.清洗"Field_02“，仅保留汉字，删除所有字母、数字于空白符 (包含换行符)
# 3.清洗"Field_03"，仅保留数字。
# 4.给”Field05“赋值，值为”Field 04“中”s“出现的次数。
# 5.删除序号为4的一行
# 6.填充”Field 06“列中的空值为"W“。将最终结果保存为同级目录下的inputr.csv


df
df=pd.read_csv('data\input.csv')
df.columns

df['Field_01'].str.count('A').sum()
df['Field_01'].str.count('B').sum()
df['Field_01'].str.count('C').sum()
df['Field_01'].str.count('D').sum()

df['Field_02']=df['Field_02'].str.split('[^\u4e00-\u9fa5]+',regex=True)
df['Field_03'].str.split('[^0-9'']+',regex=True)
df['Field_02']=df['Field_02'].str.replace('[^\u4e00-\u9fa5]+','')
df['Field_03']=df['Field_03'].str.replace('[^0-9'']+','')

df['Field_04'].str.count('s').sum()
df['Field_05']=16

df.drop(3)

df['Field_06'].str.replace('NaN','W')

df.to_csv('data\inputr.csv')
```

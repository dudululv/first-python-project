```python
import pandas as pd

# 读取数据
df = pd.read_csv('team_attributes.csv')

# 第一步：给 id 加上前缀 "a" 并设置为索引
df['id'] = 'a' + df['id'].astype(str)
df.set_index('id', inplace=True)

print("处理后的数据前5行：")
print(df.head())
print(f"\n数据形状: {df.shape}")
```

```python
import csv

FILENAME = "contacts.csv"  # ä¿®æ”¹æ–‡ä»¶æ‰©å±•å

def load_contacts():
    contacts = []
    try:
        with open(FILENAME, "r", encoding="utf-8", newline='') as f:
            reader = csv.DictReader(f)  # ä½¿ç”¨DictReaderè‡ªåŠ¨å¤„ç†è¡¨å¤´
            for row in reader:
                contacts.append({
                    "name": row["name"],
                    "phone": row["phone"], 
                    "email": row["email"]
                })
    except FileNotFoundError:
        # å¦‚æœæ–‡ä»¶ä¸å­˜åœ¨ï¼Œåˆ›å»ºå¸¦è¡¨å¤´çš„ç©ºæ–‡ä»¶
        with open(FILENAME, "w", encoding="utf-8", newline='') as f:
            writer = csv.DictWriter(f, fieldnames=["name", "phone", "email"])
            writer.writeheader()
    except Exception as e:
        print(f"âŒ è¯»å–æ–‡ä»¶æ—¶å‡ºé”™ï¼š{e}")
    return contacts

def save_contacts(contacts):
    try:
        with open(FILENAME, "w", encoding="utf-8", newline='') as f:
            writer = csv.DictWriter(f, fieldnames=["name", "phone", "email"])
            writer.writeheader()  # å†™å…¥è¡¨å¤´
            writer.writerows(contacts)  # ä¸€æ¬¡æ€§å†™å…¥æ‰€æœ‰æ•°æ®
    except Exception as e:
        print(f"âŒ ä¿å­˜æ–‡ä»¶æ—¶å‡ºé”™ï¼š{e}")

def add_contact(contacts):
    name = input("è¯·è¾“å…¥å§“åï¼š").strip()
    phone = input("è¯·è¾“å…¥ç”µè¯ï¼š").strip()
    email = input("è¯·è¾“å…¥é‚®ç®±ï¼š").strip()
    
    # æ£€æŸ¥å§“åæ˜¯å¦å·²å­˜åœ¨
    if any(c["name"] == name for c in contacts):
        print("âŒ è¯¥å§“åå·²å­˜åœ¨ï¼Œè¯·ä½¿ç”¨ä¸åŒçš„å§“åã€‚")
        return
        
    contacts.append({"name": name, "phone": phone, "email": email})
    save_contacts(contacts)  
    print(f"âœ… å·²æ·»åŠ è”ç³»äººï¼š{name}")

def search_contact(contacts):
    while True:
        print("\næŸ¥è¯¢èœå•ï¼š")
        print("1 - æŸ¥è¯¢å•ä¸ªè”ç³»äºº")
        print("2 - æŸ¥è¯¢å…¨éƒ¨è”ç³»äºº") 
        print("0 - è¿”å›ä¸»èœå•")
        choice = input("è¯·è¾“å…¥æ“ä½œç¼–å·ï¼š")

        if choice == "1":
            key = input("è¯·è¾“å…¥è¦æŸ¥è¯¢çš„å§“åæˆ–ç”µè¯ï¼š").strip()
            results = [c for c in contacts if key in c["name"] or key in c["phone"]]
            if results:
                print("æŸ¥è¯¢ç»“æœï¼š")
                for c in results:
                    print(f"å§“åï¼š{c['name']}, ç”µè¯ï¼š{c['phone']}, é‚®ç®±ï¼š{c['email']}")
            else:
                print("âŒ æœªæ‰¾åˆ°åŒ¹é…çš„è”ç³»äººã€‚")
        elif choice == "2":
            if not contacts:
                print("é€šè®¯å½•ä¸ºç©ºã€‚")
            else:
                print("ğŸ“’ æ‰€æœ‰è”ç³»äººï¼š")
                for c in contacts:
                    print(f"å§“åï¼š{c['name']}, ç”µè¯ï¼š{c['phone']}, é‚®ç®±ï¼š{c['email']}")
        elif choice == "0":
            break
        else:
            print("âŒ æ— æ•ˆè¾“å…¥ï¼Œè¯·é‡æ–°é€‰æ‹©ã€‚")

def edit_contact(contacts):
    name = input("è¯·è¾“å…¥è¦ä¿®æ”¹çš„è”ç³»äººå§“åï¼š").strip()
    for c in contacts:
        if c["name"] == name:
            print(f"å½“å‰ä¿¡æ¯ï¼šå§“åï¼š{c['name']}, ç”µè¯ï¼š{c['phone']}, é‚®ç®±ï¼š{c['email']}")
            new_phone = input("æ–°ç”µè¯ï¼ˆç•™ç©ºä¸æ”¹ï¼‰ï¼š").strip()
            new_email = input("æ–°é‚®ç®±ï¼ˆç•™ç©ºä¸æ”¹ï¼‰ï¼š").strip()
            
            if new_phone:
                c["phone"] = new_phone
            if new_email:
                c["email"] = new_email
                
            save_contacts(contacts)
            print("âœ… è”ç³»äººä¿¡æ¯å·²æ›´æ–°ã€‚")
            return
    print("âŒ æœªæ‰¾åˆ°è¯¥è”ç³»äººã€‚")

def delete_contact(contacts):
    name = input("è¯·è¾“å…¥è¦åˆ é™¤çš„è”ç³»äººå§“åï¼š").strip()
    for c in contacts:
        if c["name"] == name:
            contacts.remove(c)
            save_contacts(contacts) 
            print(f"âœ… å·²åˆ é™¤è”ç³»äººï¼š{name}")
            return
    print("âŒ æœªæ‰¾åˆ°è¯¥è”ç³»äººã€‚")

def show_all(contacts):
    if not contacts:
        print("é€šè®¯å½•ä¸ºç©ºã€‚")
    else:
        print("ğŸ“’ å½“å‰è”ç³»äººåˆ—è¡¨ï¼š")
        for c in contacts:
            print(f"å§“åï¼š{c['name']}, ç”µè¯ï¼š{c['phone']}, é‚®ç®±ï¼š{c['email']}")

# ä¸»èœå•
def main():
    contacts = load_contacts()
    print("ğŸ“± æ¬¢è¿ä½¿ç”¨è”ç³»äººç®¡ç†ç³»ç»Ÿï¼ˆCSVç‰ˆï¼Œå®æ—¶ä¿å­˜ï¼‰")

    while True:
        print("\nèœå•ï¼š")
        print("1. æ·»åŠ è”ç³»äºº")
        print("2. æŸ¥è¯¢è”ç³»äºº")
        print("3. ä¿®æ”¹è”ç³»äºº")
        print("4. åˆ é™¤è”ç³»äºº")
        print("5. æ˜¾ç¤ºå…¨éƒ¨è”ç³»äºº")
        print("0. é€€å‡ºç³»ç»Ÿ")
        choice = input("è¯·è¾“å…¥é€‰é¡¹ï¼š")

        if choice == "1":
            add_contact(contacts)
        elif choice == "2":
            search_contact(contacts)
        elif choice == "3":
            edit_contact(contacts)
        elif choice == "4":
            delete_contact(contacts)
        elif choice == "5":
            show_all(contacts)
        elif choice == "0":
            print("ğŸ“‚ ç¨‹åºå·²é€€å‡ºï¼Œæ‰€æœ‰ä¿®æ”¹å·²ä¿å­˜ã€‚")
            break
        else:
            print("âŒ æ— æ•ˆè¾“å…¥ï¼Œè¯·é‡æ–°é€‰æ‹©ã€‚")

if __name__ == "__main__":
    main()
```

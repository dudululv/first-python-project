```python
import numpy as np
import pandas as pd
```

```python
#1球员身高的Series
a=['188cm','180cm','178cm','189cm','191cm','180cm',np.NaN]
s1=pd.Series([i for i in a],index=[i for i in range(1,8)])
s1
```

```python
#2不为空的球员个数
s1.count()
```

```python
#3身高数据去掉cm
s1=s1.str.replace('[^0-9]+','',regex=True)
s1
```

```python
#4 180身高的球员个数
s2 = s1.dropna().astype(int)
count_180 = (s2 == 180).sum()
print(count_180)
```

```python
#5 处理身高数据为数值类型
s2.dtype
```

```python
#6统计185以上身高的球员个数
s2[s2>185].count()
'=='*30
```

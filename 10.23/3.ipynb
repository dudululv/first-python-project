```python
import importlib
import subprocess
import sys

package_name = "xlwt"

try:
    # 尝试导入 xlwt
    importlib.import_module(package_name)
    print(f"✅ '{package_name}' 已安装并可正常使用！")
except ImportError:
    # 如果没安装则自动安装
    print(f"⚙️ 未检测到 '{package_name}'，正在安装中...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])
    print(f"✅ '{package_name}' 安装完成！")

# 导入模块并测试
import xlwt

# 创建一个简单的 Excel 文件
workbook = xlwt.Workbook()
sheet = workbook.add_sheet("示例Sheet")

# 写入数据
sheet.write(0, 0, "姓名")
sheet.write(0, 1, "年龄")
sheet.write(1, 0, "Gump")
sheet.write(1, 1, 18)

# 保存文件
workbook.save("示例输出.xls")
print("📁 文件 '示例输出.xls' 已生成！")
```
